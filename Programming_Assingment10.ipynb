```python
# 1.	Write a Python program to find sum of elements in list?
L = [1,2,3,4,5]
addn = 0
for i in L:
  addn +=  i
print(f"Sum if above list is: {addn}")
```

```python
# 2.	Write a Python program to  Multiply all numbers in the list?
L = [1,2,3,4,5]
mult = 1
for i in L:
  mult *=  i
print(f"Multiplication of above list is: {mult}")
```

```python
# 3.	Write a Python program to find smallest number in a list?
L = [1,2,3,4,5]
# Option 1
smallest = L[0]
for i in L:
  if i<smallest:
    smallest = i  
print(f"Smallest number in above list is: {smallest}")
# Option 2
print(f"Smallest Number is: {min(L)}")
```

```python
# 4.	Write a Python program to find largest number in a list?
L = [1,2,3,4,5,9,6]
# Option 1
largest = 0
for i in L:
  if i>largest:
    largest = i
print(f"Largest Number is: {largest}")
# Option 2
print(f"largest Number is: {max(L)}")
```

```python
# 5.	Write a Python program to find second largest number in a list?
L = [1,2,3,4,5,9,6]
largest = max(L)
second_largest = 0
for i in L:
  if i != largest and i > second_largest:
    second_largest = i
print(f"Second Largest Number is {second_largest}")
```

```python
# 6.	Write a Python program to find N largest elements from a list?
L = [1,2,3,4,9,5,3,8,6,7,4,5]
N = 5
L1 = []
for i in range(0,N):
  mx = 0
  for j in L:
    if j>mx:
      mx = j
  L1.append(mx)
  L.remove(mx)
print("N Largest Numbers are:\n", L1)
```

```python
# 7.	Write a Python program to print even numbers in a list?
L = [1,2,3,4,5,9,6]
print("Even Numbers are:")
for i in L:
  if i%2==0:
    print(i)
```

```python
# 8.	Write a Python program to print odd numbers in a List?
L = [1,2,3,4,5,9,6]
print("Odd Numbers are:")
for i in L:
  if i%2 != 0:
    print(i)
```

```python
# 9.	Write a Python program to Remove empty List from List?
L = [ [1,2,3,4], [5,6,7,8], [99,22], [], [500,1233,424123]]
L1 = []
for i in L:
  if len(i) > 0:
    L1.append(i)
print(f"list after Removing Empty List is: \n {L1} ")
```

```python
# 10.	Write a Python program to Cloning or Copying a list?
L = [ [1,2,3,4], [5,6,7,8], [99,22], [], [500,1233,424123]]
L1 = []
for i in L:
  L1.append(i)
print(f"Original List is:\n {L}\n Copied List is:\n {L1}  ")
```

```python
# 11.	Write a Python program to Count occurrences of an element in a list?
L = [1,2,3,3,3,3,1,1,1,1,1,1,1,2,2,2,4,5,5,5,5]
L1 = []
for i in L:
  if i not in L1:
    L1.append(i)
print("Count of:\n")
for j in L1:
  print(f"{j} is {L.count(j)} times")
```
