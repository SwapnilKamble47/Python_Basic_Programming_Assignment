```python
# 1.	Write a Python Program to Find the Factorial of a Number?
# n! = n × (n-1) × (n-2) × (n-3) × ….× 3 × 2 × 1
A = int(input("Enter Number to check Factorial: "))

fact = 1
for i in range(1, A+1):
    fact *= i
print(f"Factorial is: ", fact)
```

```python
# 2.	Write a Python Program to Display the multiplication Table?
A = int(input("Enter Size of multiplication table"))
for i in range (1,A+1):
  print(f"Table of {i} is: ")
  for j in range (1,11):
    print(i*j)
  print("\n")
```

```python
# 3.	Write a Python Program to Print the Fibonacci sequence?
A = int(input("Enter Length of Fibonacci Sequence: "))
l = [0,1]
for i in range(1,A-1):
  fib = l[i] + l[i-1]
  l.append(fib)
print(l)
```

```python
# 4.	Write a Python Program to Check Armstrong Number?
number = int(input("Please Enter Number To Check Armstrong Number: "))

A = number
power = len(str(number))

sum = 0
while A > 0:
  temp = 0
  temp1 = 0
  temp = A % 10
  temp1 = temp**power
  sum = sum + temp1
  A = A//10
if sum == number:
  print(f"Entered Number {number} is Armstrong Number:")
else:
  print((f"Entered Number {number} is not Armstrong Number:"))

   
```

```python
# 5.	Write a Python Program to Find Armstrong Number in an Interval?
low = int(input("Enter start point of Interval: "))
high = int(input("Enter End point of Interval: "))
print(f"Armstrong numbers in range {low} to {high} are")
for i in range(low, high+1):
  power = len(str(i))
  temp = 0
  temp1 = 0
  A = i
  sum = 0
  while A > 0:  
    temp = A % 10
    temp1 = temp**power
    sum = sum + temp1
    A = A//10
  if sum == i:
     print(i)
```

```python
# 6.	Write a Python Program to Find the Sum of Natural Numbers?
A = int(input("Please Enter Natural Number to find it's Sum: "))
if A > 0:
  temp = A
  sum = 0
  while temp > 0:
    sum = sum + temp
    temp = temp-1
  print(f"Sum of Entered  Natural Number {A} is {sum}")
else:
  print(f"You Entered {A} is not Natural Number  ")
```
